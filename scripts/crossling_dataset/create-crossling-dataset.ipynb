```python
import pandas as pd

def create_dataset(path: str, from_english: bool):
    df = pd.read_parquet(path)
    options = ["Option A", "Option B", "Option C", "Option D", "Option E"]
    if from_english:
        df = df.drop(columns=["question"] + ["en_" + e for e in options])
    else:
        df = df.drop(columns=["en_question"] + options)
    return df
```

```python
df = create_dataset(
    path=r"C:\Users\nikol\OneDrive\Desktop\project_repos\CMLEval\data\sanity_check\ru_kaz\translated_KazMMLU.parquet.gzip",
    from_english=True,
)
df.insert(7, "Question", df.pop("en_question"))
df.head()
```

```python
df.to_parquet(
    r"C:\Users\nikol\OneDrive\Desktop\project_repos\CMLEval\data\crosslingual\enQ_xA\KazMMLU.parquet.gzip",
    index=False,
    compression="gzip",
)
```

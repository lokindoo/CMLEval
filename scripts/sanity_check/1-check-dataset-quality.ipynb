```python
import os 

path = r'C:\Users\nikol\OneDrive\Desktop\project_repos\CMLEval\data\sanity_check\ru_kaz\split'
dfs = []
for f in os.listdir(path):
    if f.endswith(".json"):
        print(f)
        d = pd.read_json(os.path.join(path, f))
        dfs.append(d)
dfs = pd.concat(dfs)
print(len(dfs))
dfs.head()
```

```python
dfs.insert(1, "subject", dfs["dataset_name"].apply(lambda x: x.split("_")[-1]))
dfs["dataset_name"] = dfs["dataset_name"].apply(lambda x: x.split("_")[0])
dfs
```

```python
def questions_from_dataset(subject: str) -> pd.DataFrame:
    ds = load_dataset("MBZUAI/KazMMLU", subject, split='test')
    ds = ds.to_pandas()
    ds = ds[["Question", "Level", "Option A", "Option B", "Option C", "Option D", "Option E", "Answer Key", "Language"]].copy()

    return ds
```

```python
dfs_options = pd.concat([questions_from_dataset(subject=subject) for subject in dfs["subject"].unique()])
print(len(dfs_options))
```

```python
dfs_options["Question"] = dfs_options["Question"].apply(lambda x: x.strip())
dfs_options.head()
```

```python
from tqdm import tqdm

step = 0
rows = []
for i, row in tqdm(dfs_options.iterrows(), total=len(dfs_options)):
    if step >= len(dfs):
        break
    elif row["Question"].strip() == dfs.iloc[step]["question"].strip():
        rows.append(row)
        step += 1
rows = pd.DataFrame(rows)
print(len(rows) == len(dfs))

final_df = pd.concat([dfs, rows], axis=1)
# final_df = dfs.merge(rows, left_on="question", right_on="Question", how="left")
print(len(final_df))
final_df.head()
```

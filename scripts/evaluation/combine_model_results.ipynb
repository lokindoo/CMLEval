```python
from utils.dataframe_utils import dict_to_df
from utils.io import load_json, save_df
```

```python
import pandas as pd

base_path = r"C:\Users\nikol\OneDrive\Desktop\project_repos\CMLEval\data\results\MCQA"

dfs = []
for f in os.listdir(base_path):
    if f.startswith("KazMMLU") and f.endswith(".json"):
        data = load_json(
            os.path.join(base_path, f),
        )
        data = dict_to_df(data)
        if "Direction" not in data.columns:
            if "en_q" in f:
                data["Direction"] = "en_question"
            elif "en_a" in f:
                data["Direction"] = "en_answer"
        dfs.append(data)
dfs = pd.concat(dfs)
dfs
```

```python
for model in dfs["model"].unique():
    subset = dfs[dfs["model"] == model].copy()
    correct = subset["ground_truth"] == subset["extracted_answer"]
    print(f"{model} {round(sum(correct) * 100 / len(correct), 2)}")
```
